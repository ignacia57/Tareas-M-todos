Consideremos una distribución gaussiana de dos variables $x$, $y$ dada por

$$
\rho(x,y) = \frac{\sqrt{c_{xx}c_{yy} - c_{xy}^2}}{2\pi}\exp\left[-\frac{1}{2}c_{xx} (x - \mu_x)^2 - c_{xy} (x - \mu_x)(y - \mu_y) - \frac{1}{2} c_{yy} (y - \mu_y)^2\right]
$$

La probabilidad de obtener un valor dado de $x$ independiente del valor de $y$ está dada por

$$
\rho(x) = \int_{-\infty}^\infty dy\,\rho(x,y)\,,
$$

porque sumamos sobre todos los posibles valores de $y$. A esto se lo llama "marginalizar sobre y".

Considere el caso $c_{xx} = 2$, $c_{yy} = 1$, $c_{xy} = 1.4$, $\mu_x = 3$, $\mu_y = 4$.

* Grafique las distribuciones de probabilidad $\rho(x)$ y $\rho(y)$.

* Grafique la distribución en 3 dimensiones (el eje $z$ es el valor de $\rho$).

* Suponga que esta probabilidad representa el resultado de un experimento que a partir de ciertos datos asigna una probabilidad a las cantidades $x$ y $y$. Discuta qué se puede concluir sobre el punto $x = 0$, $y = 0$.

```python
#Exportamos los paquetes necesarios para realizar la tarea
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import comb
```

Calculamos los valores de x e y, obteniendo la derivada de x y reemplazando:

$$
\rho(x) = \frac{\exp[\frac{-(x - 3)^2}{50}]}{5\sqrt{2\pi}}
$$

el valor de sigma es:
$$ 
\sigma(x) = 5
$$

Calculamos los valores de x e y, obteniendo la derivada de y, y reemplazando:

$$
\rho(y) = \frac{\exp[\frac{-(y - 4)^2}{100}]}{\sqrt{50}\sqrt{2\pi}}
$$
el valor de sigma es: 
$$
\sigma(y) = \sqrt{50}
$$

```python
%matplotlib inline
```

```python
#Definimos la gaussiana en torno a x, m y sigma como variables 
def gauss(x, mu, sigma): 
    return np.exp(-(x - mu)**2/(2*sigma**2))/np.sqrt(2*np.pi)*sigma #Escribimos la ecuacion que queremos que nos devuelva
```

```python
x = np.linspace(-15,20,100) #Definimos los valores para x 
y = np.linspace(-15,20,100) #definimos los vales para y 
 
plt.plot(x, gauss(x, 3, 5), label="p(x)") #graficamos la funcion reemplazando los valores de u(x) y de sigma
plt.plot(y, gauss(y, 4, np.sqrt(50)), label="p(y)") #Graficamos la gaussiana de p(y) reemplazando u(y) y sigma

plt.title("Distribucion de probabilidad") #Le ponemos nombre al titulo del grafico de nuestras gaussianas
plt.legend() 
plt.show()
```

```python
from mpl_toolkits.mplot3d import axes3d #Exportamos este paquete para realizar la gráfica en 3d
```

```python
#Graficamos la funcion de probabilidad en tres dimensiones 

z = (0.1/np.pi)*np.exp(-(x - 3)**2 - 1.4*(x - 3)*(y - 4) - (1/2)*(y - 4)**2) #escribimos la ecuación reemplazando los valores del principio

fig = plt.figure()
ax1 = fig.add_subplot(111,projection='3d')

plt.title("Grafico en 3d de la funciond de distribucion") #titulo del grafico 
plt.xlabel("eje x") #Nombramos el eje
plt.ylabel("eje y") #Nombramos el otro eje 
ax1.plot(x, y, z) #definimos las variables del grafico 
plt.show()
```

Finalmente podemos concluir que el punto x = 0, y = 0 corresponde a donde la distribucion comienza a tener un mayor valor y a aumentar la probabilidad
